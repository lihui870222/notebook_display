## cluster evaporation calculation based on Eq. (5) in Gnedin, Ostriker & Tremaine 14

```python
def Mf_eva(M_i, t_dur):
    """
    The final mass, M_f, of a cluster with initial mass M_i during time t_dur
    Mass is in unit Msun, t_dur in Myr
    """
    M_f = M_i-t_dur*11.76
    if M_f>=0:
        return M_f
    else:
        return 0.0
    
def Mf_eva_discrete(M_i, t_dur, dt):
    M_f = M_i
    while t_dur>dt:
        if M_f==0:
            return 0.0
        t_iso = 1.7e4*M_f/2e5
        M_f *= max( (1-dt/t_iso), 0.0 )
        t_dur -= dt
        
    t_iso = 1.7e4*M_f/2e5
    M_f *= max( (1-dt/t_iso), 0.0 )
    return M_f

def Mf_eva_exp(M_i, t_dur, dt):
    M_f = M_i
    while t_dur>dt:
        if M_f==0:
            return 0.0
        t_iso = 1.7e4*M_f/2e5
        M_f *= np.exp(-1*dt/t_iso)
        t_dur -= dt
        
    t_iso = 1.7e4*M_f/2e5
    M_f *= np.exp(-1*dt/t_iso)
    return M_f
```

```python
print "exact:", Mf_eva(1e6, 5e4)
print "dicrete:", Mf_eva_discrete(1e6, 5e4, 1000)
print "exp:", Mf_eva_exp(1e6, 5e4, 1000)
```

## evaluate the accuracy of the destruction calculations

```python
mi = 1e6
t = 6e4
dt_plot = np.logspace(1,2,100)
mf_appro = np.array([Mf_eva_discrete(mi, t, dt) for dt in dt_plot])
mf_exp = np.array([Mf_eva_exp(mi, t, dt) for dt in dt_plot])

plt.axhline(1, c='k', ls='--')
plt.plot(dt_plot, mf_appro/Mf_eva(mi,t), label=r'$1-\Delta t/\tau$')
plt.plot(dt_plot, mf_exp/Mf_eva(mi,t),  label=r'$\exp(-\Delta t/\tau)$')
plt.xlabel(r'$\Delta t \, (\rm Myr)$')
plt.ylabel(r'$M_f^{\rm est}/M_f^{\rm true}$')
plt.legend()
```

```python
def ave_timescale(t, f):
    if f<1e-6:
        return 0
    elif f==1:
        return np.Inf
    else:
        return -1.*t/np.log(f)
```

```python
N_cluster = len(age)
tt = np.zeros(N_cluster)
for i in range(N_cluster):
    tt[i] = ave_timescale(age[i],f[i])
```

```python
def median_2D(x,y,N_bins=10, logx=True):
    m = np.zeros([N_bins-1,4])
    
    if logx:
        bins = np.logspace(np.log10(np.min(x)), np.log10(np.max(x)), N_bins)
        m[:,0] = np.sqrt(bins[:-1]*bins[1:])
    else:
        bins = np.linspace(np.min(x), np.max(x), N_bins)
        m[:,0] = (bins[:-1]+bins[1:])/2.

    for i in range(N_bins-1):
        index = (x>bins[i]) & (x<bins[i+1])
        print y[index]
        m[i,1] = np.median(y[index])
        m[i,2] = np.percentile(y[index], 25)
        m[i,3] = np.percentile(y[index], 75)
    return m
```
