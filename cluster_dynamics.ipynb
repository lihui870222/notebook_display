```python
N_halo = len(halo_ds)
x_largest = np.zeros(N_halo)
y_largest = np.zeros(N_halo)
z_largest = np.zeros(N_halo)
r_largest = np.zeros(N_halo)
for i in range(N_halo):
    x_largest[i] = halo_ds[i][1]*data_ds.arr(1, "code_length")
    y_largest[i] = halo_ds[i][2]*data_ds.arr(1, "code_length")
    z_largest[i] = halo_ds[i][3]*data_ds.arr(1, "code_length")
    r_largest[i] = halo_ds[i][-2]*data_ds.arr(1, "code_length")

# cut
print '1:', x_largest[0], y_largest[0], z_largest[0], r_largest[0]
print '2:', x_largest[1], y_largest[1], z_largest[1], r_largest[1]
print '3:', x_largest[2], y_largest[2], z_largest[2], r_largest[2]
i = 0
halo_center = [x_largest[i], y_largest[i], z_largest[i]]
sp = data_ds.sphere(halo_center, r_largest[i])
```

```python
def ave_timescale(t, f):
    if f<1e-6:
        return 0
    elif f==1:
        return np.Inf
    else:
        return -1.*t/np.log(f)
```

```python
N_cluster = len(age)
tt = np.zeros(N_cluster)
for i in range(N_cluster):
    tt[i] = ave_timescale(age[i],f[i])
```

```python
def median_2D(x,y,N_bins=10, logx=True):
    m = np.zeros([N_bins-1,4])
    
    if logx:
        bins = np.logspace(np.log10(np.min(x)), np.log10(np.max(x)), N_bins)
        m[:,0] = np.sqrt(bins[:-1]*bins[1:])
    else:
        bins = np.linspace(np.min(x), np.max(x), N_bins)
        m[:,0] = (bins[:-1]+bins[1:])/2.

    for i in range(N_bins-1):
        index = (x>bins[i]) & (x<bins[i+1])
        print y[index]
        m[i,1] = np.median(y[index])
        m[i,2] = np.percentile(y[index], 25)
        m[i,3] = np.percentile(y[index], 75)
    return m
```
