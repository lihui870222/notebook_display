```python
from yt import derived_field

@derived_field(name = ('artio', 'e_turb'), units = "code_mass/code_length/code_time/code_time")
def Eturb(field, data):
    return data[('artio', 'HVAR_GAS_TURBULENT_ENERGY')]*data.ds.arr(1.0, "code_mass/code_length/code_time/code_time")

@derived_field(name = ('artio', 'v_turb'), units = "km/s")
def Vturb(field, data):
    return np.sqrt(2.0*data[('artio', 'e_turb')]/data[('gas', 'density')])
```
