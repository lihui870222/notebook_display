## read in simulation and relevant halo catalog

```python
data_file = './OUT/continuous_a0.1302.art'
data_ds = yt.load(data_file)
```

```python
ad = data_ds.all_data()
```

```python
mass = ad[("STAR", "MASS")].in_units("Msun")
age = ad[("STAR", "age")].in_units("Myr")

star_ctime = data_ds.current_time.in_units('Myr')-age
```

```python
data_ds.field_list
```

```python
ZZ = ad[('STAR', 'AVERAGE_LEVEL')]
Z = ad[('STAR', 'METALLICITY_SNII')]
```

```python
d = ZZ/mass.in_units('code_mass').value - Z*Z
```

```python
std = np.sqrt(d[d>0])
```

```python
plt.hist(std)
```

```python
plt.scatter(Z[d>0], std, s=1, alpha=0.8)
plt.plot([1e-4,1],[1e-4,1], label='$Z=\sigma_Z$')
plt.xlim(1e-4,0.02)
plt.ylim(1e-4,0.02)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$M_Z/M$')
plt.ylabel(r'$\sigma_Z$')
plt.legend()
```

```python
plt.scatter(mass[d>0], std, s=1, alpha=0.8)
plt.xlim(1e3,1e5)
plt.ylim(1e-4,2e-2)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$M \; (M_\odot)$')
plt.ylabel(r'$\sigma_Z$')
```
