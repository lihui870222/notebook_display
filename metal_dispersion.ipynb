## read in simulation and relevant halo catalog

```python
data_file = './OUT/continuous_a0.1302.art'
data_ds = yt.load(data_file)
```

```python
ad = data_ds.all_data()
```

```python
mass = ad[("STAR", "MASS")].in_units("Msun")
age = ad[("STAR", "age")].in_units("Myr")

star_ctime = data_ds.current_time.in_units('Myr')-age
```

```python
data_ds.field_list
```

```python
ZZ = ad[('STAR', 'AVERAGE_LEVEL')]
Z = ad[('STAR', 'METALLICITY_SNII')]
ZIa = ad[('STAR', 'METALLICITY_SNIa')]
```

```python
def sf_ave_time(data, region):
    # calculate the average cluster formation time in unit of Myrs
    return YTArray(data._handle.tphys_from_tcode_array(region[("STAR", "AVERAGE_AGE")]+region[("STAR", "BIRTH_TIME")])/1e6, "Myr") - region[("STAR", "creation_time")].in_units("Myr")
```

```python
t_ave = sf_ave_time(data_ds, ad)
```
